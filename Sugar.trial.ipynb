```python
import yfinance as yf 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

#Get historical data for sugar#11
sugar = yf.Ticker('SB=F')
df_sugar = sugar.history(period="6y") #Fetch last 6 years of sugar data 

#display the first few rows
display(df_sugar.head())
```

```python
#getting multiple commodities at once
commodity_tickers = {
    'Sugar': 'SB=F',
    'Coffee': 'KC=F',
    'Cocoa': 'CC=F',
    'Cotton': 'CT=F',
    'Wheat': 'ZW=F',
    'Corn': 'ZC=F',
    'Soybeans': 'ZS=F',
}

#Fetch historical data 
commodity_data ={}
for name, ticker in commodity_tickers.items():
    commodity_data[name] = yf.Ticker(ticker).history(period="6y")["Close"]

#convert to data frame 
df_commodities = pd.DataFrame(commodity_data)

print(df_commodities.shape)
display(df_commodities.head())
```

```python
#create a simple vissualisation of the data 

import plotly.express as px
#convert index in column for plotly compatibility
df_commodities_reset = df_commodities.reset_index()

#create interactive plotly line chart 
fig = px.line(
    df_commodities_reset, x='Date', y=df_commodities_reset.columns[1:],
    title='Commodity Prices Over Time',
    labels ={'value': 'Price(USD)', 'Date': 'Date'},  
    template='plotly_white'   
)
#Show the interactive plot
fig.show()

#video to help me https://www.youtube.com/watch?v=eGWjRwqnI7E&t=200s
```
